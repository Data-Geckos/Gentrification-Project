```python
searchterms_df = pd.read_csv("ZIP_Codes_and_Postal_Cities.csv")
searchterms_df.head()
```

```python
searchterms_df = searchterms_df[[
                        "ZIP Code", 
                        "Postal City 1", 
                        "Location"
                    ]]
```

```python
# set up additional columns to hold information
searchterms_df['Starbucks Count'] = ""
searchterms_df.head()
```

```python
# Write a function that cleans the rows of the location 
def clean(pin):
#     print(pin)
    pin = pin[6:-1]
#     print(pin)
    return pin
```

```python
searchterms_df['Location']
```

```python
searchterms_df['Location'] = searchterms_df['Location'].apply(clean)
```

```python
assert clean("90713(33.84871142900005, -118.11357922799999)") == "33.84871142900005, -118.11357922799999"
```

```python
storeloc_df = pd.read_csv("store_data.csv")
storeloc_df.head()
```

```python
starbucks_df = storeloc_df.loc[storeloc_df['Country'] == 'US']
```

```python
starbucks_df.head()
```

```python
def cleanzips(postcode):
    postcode = str(postcode)
    postcode = postcode[0:5]
    return postcode
```

```python
cleanzips('99515')
```

```python
assert cleanzips("995042300") == "99504"
starbucks_df['Postcode']
```

```python
starbucks_df['ZIP Code'] = starbucks_df['Postcode'].apply(cleanzips)
```

```python
starbucks_df = starbucks_df.set_index(keys='ZIP Code')
```
