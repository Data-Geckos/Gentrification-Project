```python
%matplotlib inline

import pathlib
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(pathlib.Path("Zip_Zhvi_AllHomes.csv"))
df
```

```python
la_df = df.loc[df['City']== "Los Angeles"]
la_df
```

```python
df_clean = la_df[['RegionName', "2015-01-31", "2016-01-31", "2017-01-31", "2018-01-31", "2019-01-31", "2020-01-31" ]]
df_clean = df_clean.rename(columns={'RegionName': "Zipcode"})
df_clean
```

```python
df_clean["2015-2016 change"] = (df_clean["2016-01-31"] - df_clean["2015-01-31"]) / df_clean["2015-01-31"]
df_clean["2016-2017 change"] = (df_clean["2017-01-31"] - df_clean["2016-01-31"]) / df_clean["2016-01-31"]
df_clean["2017-2018 change"] = (df_clean["2018-01-31"] - df_clean["2017-01-31"]) / df_clean["2017-01-31"]
df_clean["2018-2019 change"] = (df_clean["2019-01-31"] - df_clean["2018-01-31"]) / df_clean["2018-01-31"]
df_clean["2019-2020 change"] = (df_clean["2020-01-31"] - df_clean["2019-01-31"]) / df_clean["2019-01-31"]
df_clean
```

```python
df_changes = df_clean[[ "Zipcode", "2015-2016 change", "2016-2017 change", "2017-2018 change", "2018-2019 change", "2019-2020 change",]]
df_changes = df_changes.set_index("Zipcode")
df_changes_transposed = df_changes.transpose()
df_changes_transposed
```

```python
df_changes_transposed.var().sort_values(ascending = False)
```

```python
df_changes_transposed.var().sort_values(ascending = False).plot(kind = "bar", title = "Variance in Year to Year percent increase over past five years")
```

```python
df_changes_transposed.().sort_values(ascending = False).plot(kind = "bar")
```

```python
df_allmonths = la_df.drop(columns = ["SizeRank", "RegionID", "RegionType", "StateName", "State", "City", "Metro", "CountyName"])
df_allmonths = df_allmonths.rename(columns={'RegionName': "Zipcode"})
df_allmonths = df_allmonths.set_index("Zipcode")
df_allmonths
```

```python
df_allmonths_changes = pd.DataFrame()
for x in range(len(df_allmonths.columns)-1):
    df_allmonths_changes[f"percent change {x}"] = (df_allmonths.iloc[:,(x+1)] - df_allmonths.iloc[:,(x)]) / df_allmonths.iloc[:,(x)]
df_allmonths_changes
```

```python
df_allmonths_changes
```

```python
allmonths_changes_handler_df = df_allmonths_changes
new_df = pd.DataFrame()
for i in range(len(df_allmonths_changes.columns)-1, 0, -1):
    new_df[f"{i}"] = allmonths_changes_handler_df.transpose().var()
    allmonths_changes_handler_df = allmonths_changes_handler_df.drop(columns = f"percent change {i}")
new_df
```

```python
new_df
```

```python

mean_variance = new_df.mean()
mean_variance
handle = new_df.transpose()

blist = pd.Series(0, index = new_df.index)
    
for i in range(len(new_df)): 
    for j in range(len(mean_variance)):
        blist.iloc[i] += new_df.iloc[i,j] - mean_variance[j]
        
blist = blist.sort_values(ascending = False)
blist.plot(kind = "bar", figsize = (20, 10))
```
