```python
%matplotlib inline

import pathlib
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(pathlib.Path("Resources/Zip_Zhvi_AllHomes.csv"))
df
```

```python
la_df = df.loc[df['City']== "Los Angeles"]
la_df
```

```python
df_allmonths = la_df.drop(columns = ["SizeRank", "RegionID", "RegionType", "StateName", "State", "City", "Metro", "CountyName"])
df_allmonths = df_allmonths.rename(columns={'RegionName': "Zipcode"})
df_allmonths = df_allmonths.set_index("Zipcode")
df_allmonths
```

```python
df_allmonths_changes = pd.DataFrame()
for x in range(len(df_allmonths.columns)-1):
    df_allmonths_changes[f"percent change {x}"] = (df_allmonths.iloc[:,(x+1)] - df_allmonths.iloc[:,(x)]) / df_allmonths.iloc[:,(x)]
df_allmonths_changes
```

```python
df_allmonths_changes.transpose().var()
```

```python
allmonths_changes_handler_df = df_allmonths_changes
new_df = pd.DataFrame()
for i in range(len(df_allmonths_changes.columns)-1, 0, -1):
    new_df[f"{i}"] = allmonths_changes_handler_df.transpose().var()
    allmonths_changes_handler_df = allmonths_changes_handler_df.drop(columns = f"percent change {i}")
new_df
```

```python
ax = new_df.transpose().plot(figsize = (20, 10), legend = None, title = "variance via set size of all years decreasing as 2020 approaches 1987")
ax.set_xlim(-25, 290)
ax.plot()
plt.savefig("variance via set size of all years decreasing as 2020 approaches 1987")
```

```python
mean_variance = new_df.mean()
mean_variance


historically_influential = pd.Series(0, index = new_df.index)
    
for i in range(len(new_df)): 
    for j in range(len(mean_variance)):
        historically_influential.iloc[i] += new_df.iloc[i,j] - mean_variance[j]
        
blist = historically_influential.sort_values(ascending = False).dropna()
blist.plot(kind = "bar", figsize = (20, 10), title = "Influence on housing price changes by zipcode")
plt.savefig("Influence on housing price changes by zipcode")
```

```python
current_variance = new_df.iloc[:,1]
current_variance.dropna().sort_values(ascending = False).plot(figsize = (20, 10), kind = "bar")
```

```python
latest_variance_over_historical_influence = current_variance / historically_influential
latest_variance_over_historical_influence.dropna().sort_values(ascending = False).plot(figsize = (20, 10), kind = "bar", title = "Candidates for Gentrification")
plt.savefig("Candidates for Gentrification")
```
